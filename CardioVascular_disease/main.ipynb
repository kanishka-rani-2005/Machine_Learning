# CARDIOVASCULAR DISEASE PREDICTION

```python
import pandas as pd;
import numpy as np;
from sklearn.preprocessing import StandardScaler;
from sklearn.model_selection import train_test_split;
from sklearn.linear_model import LogisticRegression;
from sklearn.metrics import accuracy_score;
```

## Data Colllection and Preprocessing


```python
dataframe=pd.read_csv('./cardio_data_processed.csv')
```

```python
dataframe.head()
```

```python
changes={
    'Normal':0,
    'Hypertension Stage 1':1,
    'Hypertension Stage 2':2,
    'Elevated':3,
    'Hypertensive Crisis':4
}
dataframe['bp_category']=dataframe['bp_category'].map(changes)
dataframe['bp_category_encoded']=dataframe['bp_category_encoded'].map(changes)
```

```python
dataframe.head()
```

```python
dataframe.tail()
```

```python
dataframe.info()
```

```python
dataframe.describe()
```

```python
dataframe.shape
```

```python
dataframe.isnull().sum()
```

```python
dataframe["gender"] = dataframe["gender"].astype(int)  # Already numerical
dataframe["cholesterol"] = dataframe["cholesterol"].astype(int)
dataframe["gluc"] = dataframe["gluc"].astype(int)
dataframe["smoke"] = dataframe["smoke"].astype(int)
dataframe["alco"] = dataframe["alco"].astype(int)
dataframe["active"] = dataframe["active"].astype(int)
```

```python
mapped={
    1:0,
    2:1
}

dataframe['gender']=dataframe['gender'].map(mapped)
```

```python
dataframe["cardio"].value_counts()
```

```python
dataframe.groupby('cardio').mean()
```

## Splitting the Features and Labels 

```python
X=dataframe.drop('cardio',axis=1)
Y=dataframe['cardio']
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

## STANDARDIZE THE DATA 

```python
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

## SPLITTING INTO TRAINED AND TESTED DATA 

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

## Building the Model

```python
regression=LogisticRegression()
```

```python
regression.fit(X_train,Y_train)
```

## EVALUATE THE MODEL

### ACCURACY SCORE FOR TRAINED DATA



```python
X_train_predict=regression.predict(X_train)
```

```python
prediction1=accuracy_score(X_train_predict,Y_train)
```

```python
print("Accuracy score for trained data : ",prediction1)
```

### Accuracy score for Tested Data 

```python
X_test_predict=regression.predict(X_test)
```

```python
prediction2=accuracy_score(X_test_predict,Y_test)
```

```python
print("Accuracy score for Tested data : ",prediction2)
```
