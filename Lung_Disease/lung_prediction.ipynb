# LUNG CANCER PREDICTION


## Using Logistic Regression

```python
import numpy as np;
import pandas as pd;
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Data Collection and Preprocessing 

```python
dataframe=pd.read_csv('survey lung cancer.csv')
dataframe.head()
```

```python
dataframe['LUNG_CANCER']=np.where(dataframe['LUNG_CANCER']=='YES',1,0)
dataframe.head()
```

```python
dataframe['GENDER']=np.where(dataframe['GENDER']=='M',1,0)
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe.isnull().sum()
```

```python
dataframe.describe()
```

```python
dataframe.groupby('LUNG_CANCER').mean()
```

```python
dataframe['LUNG_CANCER'].value_counts()
```

## Splitting Features and Label


```python
X=dataframe.drop('LUNG_CANCER',axis=1)
Y=dataframe['LUNG_CANCER']
```

```python
X.head()
```

```python
Y.head()
```

## Data Standardization

```python
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

## SPLITTING DATA INTO TRAIN DATA AND TEST DATA 

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
```

## Model training 

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

## Model Evaluation

### Predict accuracy of Trained data

```python
X_train_pred=model.predict(X_train)

# compare 

prediction1=accuracy_score(X_train_pred,Y_train)
print("Accuracy of predicted data : ",prediction1)
```

### Predict accuracy of Test Data

```python
X_test_pred=model.predict(X_test)

# compare 

prediction2=accuracy_score(X_test_pred,Y_test)
print("Accuracy of predicted data : ",prediction2)
```
