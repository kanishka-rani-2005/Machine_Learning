# Credit Card Fraud Detection


## Using Logistic Regression 
### Because it is a Classification Problem 

```python
import numpy as np;
import pandas as pd;
from sklearn.metrics import accuracy_score;
from sklearn.model_selection import train_test_split;
from sklearn.linear_model import LogisticRegression;
from sklearn.preprocessing import StandardScaler;
```

## DATA COLLECTION AND PREPROCESSING

```python
dataframe=pd.read_csv('./creditcard.csv')
```

#### This dataset is highly unbalanced 
###### 0 has higher values as compared to 1 so we can't feed this data directly to train model

0-> Normal Transaction
1->Fraudlent Transaction

```python
dataframe['Class'].value_counts()
# print(dataframe['Class'].value_counts().sort_index())
```

## Seperating Legit and Fraud Dataset

```python
## Seperating the data for analysis
legit=dataframe[dataframe.Class==0]
fraud=dataframe[dataframe.Class==1]
```

```python
print(legit.shape,fraud.shape)
```

```python
## Statistical Measure of this data
## Amount is column in dataset
legit.Amount.describe()  
```

```python
fraud.Amount.describe()
```

```python
## Compare the values of Both Transaction Legit and Fraud
dataframe.groupby('Class').mean()
```

Under-Sampling

```python
## Build sample dataset from original containing similar distribution of normal transaction and fraudlent transaction
## The number of fraudlent transactions -> 492
## The number of normal transactions -> 284315
```

```python
legit_sample=legit.sample(n=492)  ## take random sample from original legit of size 492
```

```python
## Concate both
new_dataset=pd.concat([legit_sample,fraud],axis=0) ## row wise
new_dataset.head()
```

```python
new_dataset.groupby('Class').mean()
```

### UNIFORM DATA 

```python
new_dataset['Class'].value_counts()
```

Mean value of unbalanced dataset and balnaced dataset is almost similar 
So we can say it is a good dataset

## Splitting Features and Targets

```python
X=new_dataset.drop('Class',axis=1)
Y=new_dataset['Class']
```

```python
X.head()
```

```python
Y.head()
```

## Standardize the data 

```python
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
X
```

## Splitting the Data into Training and Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)

```

## Building Model

```python
regression=LogisticRegression()
```

```python
regression.fit(X_train,Y_train)
```

## Evaluating the model

### Finding Accuarcy score for Trained data

```python
X_train_pred=regression.predict(X_train)

prediction1=accuracy_score(X_train_pred,Y_train)
```

```python
print('Prediction Score : ',prediction1)
```

```python
X_test_pred=regression.predict(X_test)

prediction2=accuracy_score(X_test_pred,Y_test)
```

```python
print('Prediction Score : ',prediction2)
```
